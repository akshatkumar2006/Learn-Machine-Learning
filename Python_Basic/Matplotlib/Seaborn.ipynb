# ðŸ“˜ Seaborn Complete Tutorial (Python)

Seaborn is a high-level data visualization library built on top of Matplotlib. It provides a more intuitive API and integrates well with pandas DataFrames.

---

## âœ… Getting Started
```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

---

## ðŸ“Š Datasets
Seaborn comes with several built-in datasets:
```python
sns.get_dataset_names()  # List of built-in datasets

df = sns.load_dataset("tips")
df.head()
```

---

## ðŸ“¦ 1. BarPlot
```python
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Average Total Bill by Day")
plt.show()
```

---

## ðŸ“¦ 2. HistPlot
```python
sns.histplot(df["total_bill"], kde=False, bins=20)
plt.title("Histogram of Total Bills")
plt.show()
```

---

## ðŸ“¦ 3. KDE Plot
```python
sns.kdeplot(data=df, x="total_bill", fill=True)
plt.title("KDE Plot of Total Bill")
plt.show()
```

---

## ðŸ“¦ 4. ScatterPlot
```python
sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex")
plt.title("Tip vs Total Bill")
plt.show()
```

---

## ðŸ“¦ 5. LinePlot
```python
sns.lineplot(x="size", y="total_bill", data=df)
plt.title("Line Plot of Size vs Total Bill")
plt.show()
```

---

## ðŸ“¦ 6. JoinPlot
```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
sns.jointplot(x="total_bill", y="tip", data=df, kind="hex")
```

---

## ðŸ“¦ 7. PairPlot
```python
sns.pairplot(df, hue="sex")
plt.show()
```

---

## ðŸ“¦ 8. BoxPlot
```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

---

## ðŸ“¦ 9. ViolinPlot
```python
sns.violinplot(x="day", y="total_bill", data=df)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

---

## ðŸ“¦ 10. StripPlot
```python
sns.stripplot(x="day", y="total_bill", data=df, jitter=True)
plt.title("Strip Plot")
plt.show()
```

---

## ðŸ“¦ 11. SwarmPlot
```python
sns.swarmplot(x="day", y="total_bill", data=df)
plt.title("Swarm Plot")
plt.show()
```

---

## ðŸŽ¨ Palettes & Colormaps
```python
sns.set_palette("Set2")     # Seaborn palettes: deep, muted, bright, pastel, dark, colorblind
colors = sns.color_palette("coolwarm", 6)
sns.palplot(colors)         # Display palette
```

---

## ðŸŽ¯ Context Settings
```python

sns.set_style("whitegrid")      # white, dark, whitegrid, darkgrid, ticks
```

---

## ðŸ“ HeatMap
### Using Correlation
```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Flights Dataset Heatmap
```python
flights = sns.load_dataset("flights")
pivot_flights = flights.pivot("month", "year", "passengers")
sns.heatmap(pivot_flights, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Monthly Flights Heatmap")
plt.show()
```

---

## ðŸ§  Important Concepts & Definitions

| Term       | Description |
|------------|-------------|
| **KDE**    | Kernel Density Estimate â€“ smooth estimate of the probability distribution of a dataset |
| **Boxplot**| Shows distribution using quartiles and outliers |
| **Violin** | Combines KDE and boxplot in a single plot |
| **Pairplot** | All pairwise relationships in a dataset |
| **Jointplot** | Bivariate plot with marginal histograms/KDE |
| **Palette** | Color theme for plots |
| **Context** | Controls scale of plot elements for different uses |
| **Heatmap** | Matrix-style data visualized using color intensity |
