## ðŸ§  3. Selecting & Filtering
```python

```

## ðŸ”„ 4. Data Manipulation
```python

df.set_index('B')             # Set index
```

## ðŸ“Š 5. Aggregation & Stats
```python

df.value_counts()             # Unique value counts
```
