```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fn(X):
    return 2*X**2 - 4*X + 5

def derFn(X):
    return 4*X - 4
```

```python
X = np.linspace(-3,5,100)
y = fn(X)
```

```python
xm = -2
alpha = 0.1
xml = [xm]
```

```python
for i in range(20):
    xm = xm - alpha*(derFn(xm))
    xml.append(xm)
```

```python
plt.plot(X,y)
plt.plot(xml, fn(np.array(xml)), 'r')
plt.show()
```
